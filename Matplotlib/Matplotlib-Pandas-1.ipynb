## 3. Plotting data directly with pandas 

Matplotlib has a tight integration with pandas too.

You can directly plot from a pandas DataFrame with DataFrame.plot().

Let's see the following plots directly from a pandas DataFrame:

* Line
* Scatter
* Bar
* Hist

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
# Demo
np.random.seed(123456)

ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))

ts = ts.cumsum()

ts.plot();
```

```python
# My-version

ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2024", periods=1000))

ts = ts.cumsum()

ts.plot();
```

cumsum is short for cumulative sum or continaully adding one thing to the next and so on.

## Working with an actuall data

```python
car_sales
```

```python
# Removing extra signs and commas
car_sales = pd.read_csv('car-sales.csv')

car_sales['Price'] = car_sales['Price'].str.replace('[\$,\,\.]', '', regex=True)
car_sales
```

```python
# Removing extra zero's

car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
# Adding a Date

car_sales['Sales Date'] = pd.date_range("1/1/2024", periods=len(car_sales))
car_sales
```

```python
# Doesn't work Because it's concatenating two Strings
car_sales['Total Sale'] = car_sales['Price'].cumsum()
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date', y='Total Sales');
```
