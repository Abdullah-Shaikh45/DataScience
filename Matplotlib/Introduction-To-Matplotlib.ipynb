- `plt.plot()` is a function in the Matplotlib library used to create a plot, but it doesn't display the plot immediately. It's like preparing a painting but not showing it yet.


- On the other hand, `plt.show()` is another function in Matplotlib that actually displays the plot on your screen. It's like unveiling the painting for everyone to see.


`plt.plot()` creates the plot.

`plt.show()` displays the plot.

```python
plt.plot()
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
```

```python
x =[1, 2, 3, 4]
y = [11, 22, 33, 44]
```

A few quick things about a plot:

- `x` is the horizontal axis.

- `y` is the vertical axis.

- In a data point, `x` usually comes first, e.g. (3, 4) would be (x=3, y=4).

- The same is happens in matplotlib.pyplot.plot(), x comes before y, e.g. plt.plot(x, y).

```python
plt.plot(x, y);
```

```python
# 1st Method

fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# Another Method

fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

`fig, ax` = plt.subplots(): Creates a canvas (fig) and an axes space (ax).

`ax.plot(x, y)`: Plots your data on the axes.
    
The semicolon at the end suppresses extra output.

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
# step 0: Setting up Matplotlib

import matplotlib.pyplot as plt

# step 1: Prepare Data

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# step 2: Setting-up Plot

fig, ax = plt.subplots(figsize=(10, 10))

# step 3: plot Data
ax.plot(x, y)

# step 4: Customize the plot
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')

# step 5: save & show
fig.savefig('images/simple-plot.png')
```

```python
import os

save_dir = 'images/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

fig.savefig(save_dir + 'simple-plot.png')
```
