# Manipulating Data 

```python
import pandas as pd
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
car_sales['Make'].dtype
```

```python
# Note first Here we have to to convert it into string
car_sales['Make'].str.lower()
```

```python
# But it din't change
car_sales
```

```python
# So for changing we need to re-assing it

car_sales['Make'] = car_sales['Make'].str.lower()
car_sales
```

```python
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing
```

```python
car_sales_missing['Odometer'].mean()
```

```python
# fillna(): fill the NaN colum
# Here filling the "odometer mean" to NaN

car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean())
```

```python
# it din't changed
car_sales_missing
```

```python
car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean(),inplace = True)
```

```python
car_sales_missing
```

```python
# dropna(): drops the NaN part
car_sales_missing.dropna()
```

```python
car_sales_missing.dropna(inplace = True)
```
