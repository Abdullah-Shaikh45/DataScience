# Manipulating Data part-2

```python
# column for Series

seats_column = pd.Series([5,5,5,5,5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
# Now to avoid NaN we'll use fillna()

car_sales["Seats"].fillna(5,inplace = True)
```

```python
car_sales
```

```python
# Creating a Python List

fuel_economy = [3.5, 2.3, 3.5, 6.7, 54.4 ,2.3, 5, 7.8, 8.0, 9.0]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales['Total Fuel used (L)'] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100KM']
```

```python
car_sales
```

```python
car_sales['Number of Wheels'] = 4
car_sales
```

```python
car_sales["Passed Road Safety"] = True
car_sales.dtypes
```

```python
car_sales = car_sales.drop["Total Fuel used", axis = 1]
```

```python
# Drop the Price per KM column
car_sales = car_sales.drop("Price per KM", axis=1) # columns live on axis 1
car_sales
```
